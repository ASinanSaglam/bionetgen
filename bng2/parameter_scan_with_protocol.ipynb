```python
import sys
import matplotlib.pyplot as plt
%matplotlib inline
file = open('nfkb.gdat')
lst = []
for line in file:
    lst += [line.split()]
    
column0 = [x[0] for x in lst]
column1 = [x[1] for x in lst]
plt.plot(column0[1:],column1[1:])
```

```python
vals= [2,3.4,4,5]
for i in range(1,5):
    file = open('nfkb_illustrating_protocols_R/nfkb_illustrating_protocols_R_0000'+str(i)+'.gdat')
    lst = []
    for line in file:
        lst += [line.split()]

    column0 = [x[0] for x in lst]
    column1 = [x[1] for x in lst]
    plt.plot(column0[1:],column1[1:],label=str(10**vals[i-1]))
    plt.hold
plt.legend()
```
